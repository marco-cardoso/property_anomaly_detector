```python
import pandas as pd

from property_anomaly_detector.features.read_df import read_df
from property_anomaly_detector.features.feature_engineer import flatten_rental_prices, normalize_features
from property_anomaly_detector.ml.detect_anomalies import detect

%matplotlib inline

%load_ext autoreload
%autoreload 2

pd.set_option('display.max_rows', 700)
```

# Top 200 Outliers

```python
df = read_df(default_filter={'status' : 'to_rent'})
df = flatten_rental_prices(df)

normalize_mask = ['latitude', 'longitude', 'monthly_rental_price']

df['price'] = df['monthly_rental_price']

df[normalize_mask] = normalize_features(df[normalize_mask])
anomalies = detect(df, columns=normalize_mask)


re = anomalies[['property_type', 'furnished_state', 'shared_occupancy', 'num_bedrooms', 'num_bathrooms', 'price', 'outlier_score']].sort_values(by='outlier_score', ascending=False)
re[:200]
```
